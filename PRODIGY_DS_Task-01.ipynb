```python
import pandas as pd
import matplotlib.pyplot as plt


# Loading the dataset 
df = pd.read_csv('under_five_mortality_ethiopia.csv')

# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.barh(df['Cause of Death'], df['Percentage'], color='skyblue')
plt.xlabel('Percentage of Under-Five Mortality (%)')
plt.title('Causes of Under-Five Mortality in Ethiopia')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlim(0, 20) 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('cancer_statistics_dataset.csv')

import seaborn as sns
sns.set(style='whitegrid')

# Function to plot bar charts
def plot_bar_chart(data, title, xlabel, ylabel):
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Subcategory', y='Value', data=data, palette='viridis')
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Bar chart for Global Cancer Deaths by Type
global_deaths = df[df['Category'] == 'Global Cancer Deaths by Type']
plot_bar_chart(global_deaths, 'Global Cancer Deaths by Type', 'Cancer Type', 'Deaths')

# Bar chart for Cancer Deaths in Africa by Type
africa_deaths = df[df['Category'] == 'Cancer Deaths in Africa by Type']
plot_bar_chart(africa_deaths, 'Cancer Deaths in Africa by Type', 'Cancer Type', 'Deaths')

# Bar chart for Cancer Deaths in Ethiopia by Type
ethiopia_deaths = df[df['Category'] == 'Cancer Deaths in Ethiopia by Type']
plot_bar_chart(ethiopia_deaths, 'Cancer Deaths in Ethiopia by Type', 'Cancer Type', 'Deaths')

# Pie chart for cancer deaths in Ethiopia
plt.figure(figsize=(8, 8))
plt.pie(ethiopia_deaths['Value'], labels=ethiopia_deaths['Subcategory'], autopct='%1.1f%%', startangle=140)
plt.title('Cancer Deaths Distribution in Ethiopia', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```
